```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
# Load image
image = cv2.imread("book.jpg")
```

```
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```
# Horizontal edges
sobel_horizontal = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_horizontal = np.uint8(np.absolute(sobel_horizontal))
```

```
# Vertical edges
sobel_vertical = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
sobel_vertical = np.uint8(np.absolute(sobel_vertical))
```

```
# Combine edges
sobel_combined = cv2.addWeighted(sobel_horizontal, 0.5, sobel_vertical, 0.5, 0)
```

```
# Display results
plt.figure()
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```
plt.figure()
plt.imshow(sobel_horizontal, cmap='gray')
plt.title("Horizontal Edges")
plt.axis("off")
plt.show()
```

```
plt.figure()
plt.imshow(sobel_vertical, cmap='gray')
plt.title("Vertical Edges")
plt.axis("off")
plt.show()
```

```
plt.figure()
plt.imshow(sobel_combined, cmap='gray')
plt.title("Combined Edges")
plt.axis("off")
plt.show()
```
